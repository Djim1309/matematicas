## Práctica 3.4


```python
import numpy as np
import time

# Guarda el tiempo de inicio
start_time = time.time()

# Crea una matriz cuadrada de dimensiones 3x3 con números aleatorios entre -9 y 9, excepto 0
matrix = np.random.randint(-9, 9, (3, 3))
while np.linalg.det(matrix) == 0:
    matrix = np.random.randint(-9, 9, (3, 3))

# Imprime la matriz y su determinante si es igual a 1
if np.linalg.det(matrix) == 1:
    print("Matriz:")
    print(matrix)
    print("Determinante:", np.linalg.det(matrix))

# Calcula y muestra la transpuesta de la matriz
print("Transpuesta:")
print(matrix.T)

# Calcula y muestra la matriz inversa
inv = np.linalg.inv(matrix)
print("Inversa:")
print(inv)

# Comprueba que la matriz multiplicada por su inversa es la matriz identidad
product = np.dot(matrix, inv)
print("Producto de la matriz por su inversa:")
print(product)

# Calcula y muestra el tiempo transcurrido
elapsed_time = time.time() - start_time
print("Tiempo transcurrido:", elapsed_time, "segundos")
```

##Práctica 5


```python
import numpy as np

# Creamos la matriz A y el vector b con los coeficientes y términos independientes dados
A = np.array([[1, 3, -1, -2], 
              [4, 2, 5, -3], 
              [-2, 1, 3, 1], 
              [3, -1, 2, 4]])
b = np.array([6, 3, 5, 7])

# Imprimimos la matriz A
print("Matriz A:")
print(A)

# Calculamos y imprimimos el determinante de A
det_A = np.linalg.det(A)
print("Determinante de A:", det_A)

# Calculamos y imprimimos la matriz inversa de A
A_inv = np.linalg.inv(A)
print("Matriz inversa de A:")
print(A_inv)

# Resolvemos el sistema de ecuaciones de la forma Ax = b usando A_inv
x = A_inv @ b
print("Solución del sistema de ecuaciones (método 1):", x)

# Comprobamos que x es una solución del sistema de ecuaciones
print("Comprobación:", A @ x)

# Resolvemos el sistema de ecuaciones usando np.linalg.solve
x = np.linalg.solve(A, b)
print("Solución del sistema de ecuaciones (método 2):", x)

# Comprobamos que x es una solución del sistema de ecuaciones
print("Comprobación:", A @ x)
```
